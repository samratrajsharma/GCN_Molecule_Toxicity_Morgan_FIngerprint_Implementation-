```python
import pandas as pd
```

```python
from rdkit import Chem
from rdkit.Chem import PandasTools

supplier = Chem.SDMolSupplier('Tox21_dataset.sdf')

df = PandasTools.LoadSDF('Tox21_dataset.sdf')

print(df.head())
```

```python
df = df.dropna()
```

```python
df
```

```python
df['FW'] = df['FW'].str.extract(r'(\d+\.\d+)').astype(float)
```

```python
target_variables = ['NR-AR',
        'SR-ARE', 'NR-Aromatase', 'NR-ER-LBD', 'NR-AhR', 'SR-MMP', 'NR-ER',
        'NR-PPAR-gamma', 'SR-p53', 'SR-ATAD5', 'NR-AR-LBD'] 

X = df[['FW', 'DSSTox_CID', 'SR-HSE']]  
y = df[target_variables]  

print("Features shape:", X.shape)
print("Target shape:", y.shape)
```

```python
X.isnull().sum(), y.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.multioutput import MultiOutputClassifier
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
multi_target_model = MultiOutputClassifier(model)
multi_target_model.fit(X_train, y_train)
```

```python
y_pred = multi_target_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
for i in range(y_test.shape[1]):
    accuracy = accuracy_score(y_test.iloc[:, i], y_pred[:, i])
    print(f'Accuracy for target {y_test.columns[i]}: {accuracy:.2f}')
```
