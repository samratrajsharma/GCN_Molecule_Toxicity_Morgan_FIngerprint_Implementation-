```python
import pandas as pd
```

```python
from rdkit import Chem
from rdkit.Chem import PandasTools

supplier = Chem.SDMolSupplier('Tox21_dataset.sdf')

df = PandasTools.LoadSDF('Tox21_dataset.sdf')

print(df.head())
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df
```

```python
df['FW'] = df['FW'].str.extract(r'(\d+\.\d+)').astype(float)
```

```python
target_variables = ['NR-AR',
        'SR-ARE', 'NR-Aromatase', 'NR-ER-LBD', 'NR-AhR', 'SR-MMP', 'NR-ER',
        'NR-PPAR-gamma', 'SR-p53', 'SR-ATAD5', 'NR-AR-LBD'] 

X = df[['FW', 'DSSTox_CID', 'SR-HSE']]  
y = df[target_variables]  

print("Features shape:", X.shape)
print("Target shape:", y.shape)
```

```python
X
```

```python
y
```

```python
df
```

```python
X.isnull().sum(), y.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.multioutput import MultiOutputClassifier
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
multi_target_model = MultiOutputClassifier(model)
multi_target_model.fit(X_train, y_train)
```

```python
y_pred = multi_target_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
for i in range(y_test.shape[1]):
    accuracy = accuracy_score(y_test.iloc[:, i], y_pred[:, i])
    print(f'Accuracy for target {y_test.columns[i]}: {accuracy:.2f}')
```

```python
from sklearn.metrics import classification_report

for i, target in enumerate(y_test.columns):
    print(f"Classification Report for {target}:")
    print(classification_report(y_test.iloc[:, i], y_pred[:, i]))
    print("-" * 50)
```

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['NR-AR', 'SR-ARE', 'NR-Aromatase', 'NR-ER-LBD', 'NR-AhR', 'NR-AR-LBD']
precision = [[1.00, 0.00], [0.95, 0.00], [0.99, 0.00], [0.99, 0.00], [1.00, 0.00], [1.00, 0.00]]
recall = [[1.00, 0.00], [1.00, 0.00], [1.00, 0.00], [0.99, 0.00], [1.00, 0.00], [1.00, 0.00]]
f1_score = [[1.00, 0.00], [0.98, 0.00], [1.00, 0.00], [0.99, 0.00], [1.00, 0.00], [1.00, 0.00]]

x = np.arange(len(labels))
width = 0.35

fig, axes = plt.subplots(3, 1, figsize=(10, 12))

axes[0].bar(x - width/2, [p[0] for p in precision], width, label='Class 0', color='blue')
axes[0].bar(x + width/2, [p[1] for p in precision], width, label='Class 1', color='red')
axes[0].set_title('Precision')
axes[0].set_xticks(x)
axes[0].set_xticklabels(labels)
axes[0].legend()

axes[1].bar(x - width/2, [r[0] for r in recall], width, label='Class 0', color='blue')
axes[1].bar(x + width/2, [r[1] for r in recall], width, label='Class 1', color='red')
axes[1].set_title('Recall')
axes[1].set_xticks(x)
axes[1].set_xticklabels(labels)
axes[1].legend()

axes[2].bar(x - width/2, [f[0] for f in f1_score], width, label='Class 0', color='blue')
axes[2].bar(x + width/2, [f[1] for f in f1_score], width, label='Class 1', color='red')
axes[2].set_title('F1 Score')
axes[2].set_xticks(x)
axes[2].set_xticklabels(labels)
axes[2].legend()

plt.tight_layout()
plt.show()
```
